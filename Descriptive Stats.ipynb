**Q4: Median in Real Life (Medium)**

**A car dealer notes the prices of used cars: [$8,000, $9,500, $10,200, $11,000, $50,000].**

**Why is the median a better measure than the mean in this case? Calculate the median**

**Ans-**

**Arrange Data in Ascending Order -
[$8,000,$9,500,$10,200,$11,000,$50,000]**

There are 5 values (odd number).

**The median is the middle value- $10,500**

**Reason median is a better measure than mean**

--->

* The price $50,000 is an extreme value (outlier).

* This very high price would pull the mean upward, making it misleading.

* The median is not affected by extreme values, so it better represents the typical used car price.

*   **The median price is $10,200, and it is a better measure than the mean because it gives a more realistic idea of the usual car price in the presence of an outlier.**

**Q8: Standard Deviation in Context (Hard)**


**A finance student compares the daily price fluctuations of two cryptocurrencies.**

**Coin A: standard deviation = $30**


**Coin B: standard deviation = $120**


**Which coin is riskier to invest in, and why**

**Ans-**

**Standard deviation measures how much prices typically deviate from the average price.**


**Coin (B) is riskier to invest in.**

Why?

* Coin B’s price changes vary much more from day to day.

* Coin A is comparatively more stable with smaller daily price changes.

* **Coin B is riskier because its much higher standard deviation indicates greater price volatility and uncertainty.**
